### Уровень 0:

Посчитайте значение 1-ой и 2-ой производной фунции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$

В точке x = 1.

```
import numpy as np
from scipy.misc import derivative
import matplotlib.pyplot as plt
```

```
from sympy import diff, symbols, cos, sin
```

```
x = symbols('x')
difff_1 = diff(x**5 + 4*sin(2*x) + cos(3*x + 3)).evalf(subs={'x':1})
difff_2 = diff(diff(x**5 + 4*sin(2*x) + cos(3*x + 3))).evalf(subs={'x':1})

difff_1, difff_2
```

*Или через derivative*

```
def func(x):
    y = x**5 + 4*np.sin(x*2) + np.cos(3*x+3) 
    return y

diff_1 = derivative(func, 1, dx=1e-6)
diff_2 = derivative(func, 1, dx=1e-6, n=2)

diff_1, diff_2
```

### Уровень 1:

Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
```

```
x = Symbol('x')
y = ((sin(2*x+1))**5)
y_first = y.diff(x)
y_second = y_first.diff(x)
print('Первая производная функции: ', y_first)
print('Вторая производная функции: ', y_second)
```

```
def f1(x):
    return 10*sin(2*x + 1)**4*cos(2*x + 1)
def f2(x):
    return -20*sin(2*x + 1)**5 + 80*sin(2*x + 1)**3*cos(2*x + 1)**2
```

```
x1 = np.linspace(-5, 5, 100)
y1 = [f1(x1[i]) for i in range(len(x1))]
y2 = [f2(x1[i]) for i in range(len(x1))]  

plt.plot(x1, y1, label = 'Первая производная')
plt.plot(x1, y2, label = 'Вторая производная')
plt.title('Графики производной функции')
plt.legend()
plt.ylabel('y')
plt.xlabel('x')
plt.show()
```
